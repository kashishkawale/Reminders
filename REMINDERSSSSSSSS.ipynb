```python
import datetime
import tkinter as tk
from tkinter import messagebox
import platform
import os

class DateReminder:
    def __init__(self):
        self.reminders = {}

    def add_reminder(self, date, reminder):
        if date not in self.reminders:
            self.reminders[date] = []
        self.reminders[date].append(reminder)
        messagebox.showinfo("Success", f"Reminder added for {date}")

    def check_reminders(self):
        today = datetime.date.today()
        date = today.strftime("%Y-%m-%d")
        reminders = self.reminders.get(date, ["No reminders for today."])
        reminder_message = "\n".join(reminders)
        messagebox.showinfo("Date Reminders", reminder_message)
        self.play_notification_sound()

    def play_notification_sound(self):
        if platform.system() == "Windows":
            import winsound
            winsound.Beep(2500, 1000)  # 2500 Hz frequency, 1000 ms duration
        else:
            print("\a")  # Bell sound

def add_reminder():
    date = date_entry.get().strip()
    reminder = reminder_text.get("1.0", tk.END).strip()
    if date and reminder:
        reminder_system.add_reminder(date, reminder)
        date_entry.delete(0, tk.END)
        reminder_text.delete("1.0", tk.END)
    else:
        messagebox.showwarning("Input Error", "Please enter both date and reminder.")

def quit_app():
    root.destroy()

def main():
    global reminder_system, date_entry, reminder_text, root
    reminder_system = DateReminder()
    root = tk.Tk()
    root.title("Date Reminder System")

    tk.Label(root, text="Date (YYYY-MM-DD):").grid(row=0, column=0)
    date_entry = tk.Entry(root)
    date_entry.grid(row=0, column=1)

    tk.Label(root, text="Reminder:").grid(row=1, column=0)
    reminder_text = tk.Text(root, height=10, width=30)
    reminder_text.grid(row=1, column=1)

    button_frame = tk.Frame(root)
    button_frame.grid(row=2, column=0, columnspan=2)

    tk.Button(button_frame, text="Add Reminder", command=add_reminder).pack(side=tk.LEFT)
    tk.Button(button_frame, text="Check Reminders", command=reminder_system.check_reminders).pack(side=tk.LEFT)
    tk.Button(button_frame, text="Quit", command=quit_app).pack(side=tk.LEFT)

    root.mainloop()

if __name__ == "__main__":
    main()
```
